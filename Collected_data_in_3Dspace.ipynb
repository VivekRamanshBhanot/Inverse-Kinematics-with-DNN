```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data2 = pd.read_csv('/content/robo_data_TESTING(1).txt')
```

```python
data2.drop([' w1',' w2',' w3'],1,inplace=True)
```

```python
data2
```

```python
data2.columns = ['x','y','z','alpha','beta','gamma']
#xyz are the coordinates of the tip of the arm with servo angles as alpha, beta, gamma
```

```python
data2.describe()
```

```python
ax = plt.axes(projection='3d')
ax.scatter(data2['x'],data2['y'],data2['z'],cmap='virdis',alpha=0.01)
```

```python
# points swipped by the tip of the arm by keeping alpha constant and varying beta and gamma.
# This is a crossectional view of the points recorded when the space is cut by a vertical plane
# passing through the origin (i.e. the base of the arm)
pick = data2[data2['alpha']==120]
plt.scatter(pick['y'], pick['z'], alpha=0.1)
```
